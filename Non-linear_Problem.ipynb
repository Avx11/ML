```python
import numpy as np
from matplotlib import pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate new data: Years of Experience (X) vs. Salary (y)
np.random.seed(42)
X = np.linspace(0, 20, 100).reshape(-1, 1)  # 0-20 years of experience
y = 50 * np.log(X + 1) + 10 * np.random.randn(100, 1)  # Logarithmic salary trend + noise

# Convert to Python lists
X_list = X.flatten().tolist()  # [0.0, 0.2, 0.4, ..., 20.0]
y_list = y.flatten().tolist()  # Corresponding salaries

print("First 5 X (Years):", X_list[:5])
print("First 5 y (Salary):", [f"${val:.2f}K" for val in y_list[:5]])
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
ax.scatter(X, y,color="green")
```

```python
model = LinearRegression()
model.fit(X, y)
y_hat = model.predict(X)
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
ax.scatter(X, y,color="green")
ax.plot(X, y_hat)
```

## Polynomial regression

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import Ridge
```

```python
X_poly = PolynomialFeatures(3).fit_transform(X)
```

```python
X_poly.shape, X.shape
```

```python
model2 = Ridge(alpha=0.003)
model2.fit(X_poly, y)
```

```python
y_hat2 = model2.predict(X_poly)
```

```python
X_b = np.linspace(X.min(), X.max(), 10).reshape(-1, 1)
X_b_poly = PolynomialFeatures(3).fit_transform(X_b)
y_b_hat = model2.predict(X_b_poly)
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
ax.scatter(X, y,color="green")
ax.plot(X_b, y_b_hat, color="blue")
ax.set_ylabel("y")
ax.set_xlabel("X")
```
