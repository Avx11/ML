```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import mean_squared_error
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate new data: Years of Experience (X) vs. Salary (y)
np.random.seed(42)
X = np.linspace(0, 20, 100).reshape(-1, 1)  # 0-20 years of experience
y = 50 * np.log(X + 1) + 10 * np.random.randn(100, 1)  # Logarithmic salary trend + noise

# Convert to Python lists
X_list = X.flatten().tolist()  # [0.0, 0.2, 0.4, ..., 20.0]
y_list = y.flatten().tolist()  # Corresponding salaries

print("First 5 X (Years):", X_list[:5])
print("First 5 y (Salary):", [f"${val:.2f}K" for val in y_list[:5]])
```

```python
plt.scatter(X_list, y_list, color='pink', label='Data')
plt.title('Synthetic Dataset')
plt.xlabel('X')
plt.ylabel('y')
plt.legend()
plt.show()
```

## Linear Regression

```python
linear_model = LinearRegression()
linear_model.fit(X, y)

y_pred_linear = linear_model.predict(X)

plt.scatter(X, y, color='pink', label='Data')
plt.plot(X, y_pred_linear, color='red', label='Linear Regression')
plt.title('Linear Regression')
plt.xlabel('X')
plt.ylabel('y')
plt.legend()
plt.show()
```

## Polynomial Regression

```python
degrees = [1, 2, 5]
plt.scatter(X, y, color='pink', label='Data')

for degree in degrees:
    poly_features = PolynomialFeatures(degree=degree, include_bias=False)
    X_poly = poly_features.fit_transform(X)
    
    poly_model = LinearRegression()
    poly_model.fit(X_poly, y)
    
    y_pred_poly = poly_model.predict(X_poly)
    plt.plot(X, y_pred_poly, label=f'Degree {degree}')

plt.title('Polynomial Regression with Different Degrees')
plt.xlabel('X')
plt.ylabel('y')
plt.legend()
plt.show()
```
